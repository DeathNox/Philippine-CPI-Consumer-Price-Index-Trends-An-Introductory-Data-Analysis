```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import plotly.express as px
import seaborn as sns

%config InlineBackend.figure_format = 'retina'

import warnings
warnings.filterwarnings(action='ignore', category=FutureWarning) 
```

```python
df = pd.read_csv("df_read_transformed_partial.csv")
df.head()
```

```python
df_sliced = df.loc[:, 'Unnamed: 15':'Personal Care, and Miscellaneous Goods and Services.1']

df_sliced.to_csv('df_sliced.csv', index=False)

df_sliced = df_sliced.rename(columns={'Unnamed: 16': 'All items'})
df_sliced = df_sliced.rename(columns={'Unnamed: 15': 'Month'})
df_sliced.loc[0, 'Month'] = 'Jan'
df_sliced.loc[1, 'Month'] = 'Feb'
df_sliced.loc[2, 'Month'] = 'Mar'
df_sliced.loc[3, 'Month'] = 'Apr' 
df_sliced.loc[4, 'Month'] = 'May'
df_sliced.loc[5, 'Month'] = 'Jun'

# Remove '.1' from all column names in df_sliced
df_sliced.columns = df_sliced.columns.str.replace('.1', '', regex=False)

df_sliced.head()
  
```

```python
# Drop rows by index (for example, rows 0 to 5)
df_cut = df.drop(columns=df.loc[:, 'Unnamed: 15':'Personal Care, and Miscellaneous Goods and Services.1'].columns)
df_cut = df_cut.drop(index=[13,14])
df_cut.head()
```

```python
df_combined = pd.concat([df_cut, df_sliced], axis=0)


df_combined
```

```python
df_combined = df_combined.dropna()

df_combined.to_csv('df_combined.csv', index=False)


df_combined
```

```python
# Insert a new column 'Year' at position 1 (right after 'Month')
df_combined.insert(1, 'Year', 2025)
```

```python
df_combined = df_combined.reset_index(drop=True)
df_combined
```
