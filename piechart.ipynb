```python
import matplotlib.pyplot as plt

job_data = ['40', '30', '20', '10']
labels = ['Engineer', 'Manager', 'Sales', 'HR']
explode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Engineer')

plt.pie(job_data, explode=explode, labels=labels)

plt.show()
```
